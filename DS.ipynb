```python
X = df.drop(columns='Avg Salary(K)')
X
```

```python
df[['Job Title','Salary Estimate', 'Job Description', 'Rating', 'Company Name', 'Location' ,'Headquarters', 'Size', 'Founded','Avg Salary(K)']]
```

```python
for column in X.columns:
    if X[column].dtype == type(object):
        le = sklearn.preprocessing.LabelEncoder()
        X[column] = le.fit_transform(X[column])
X.tail()
```

```python

X = pd.DataFrame(sklearn.preprocessing.StandardScaler().fit_transform(X), columns=X.columns)
y = df['Avg Salary(K)']
```

```python
y
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=666)
```

```python

import statsmodels.api as sm
import sklearn 
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
from xgboost import XGBClassifier
```

```python
%%time
LR = LogisticRegression(solver='lbfgs', max_iter=100000, random_state=555)
RF = RandomForestRegressor(n_estimators = 100, random_state=555)
SVM = SVC(random_state=0, probability=True)
KNR = KNeighborsRegressor()
clfs = []
print('5-fold cross validation:\n')
for clf, label in zip([LR, RF, SVM,KNR],
                      ['Logistic Regression',
                       'Random Forest',
                       'Support Vector Machine',
                       'KNeighbors',
                       ]):
    scores = sklearn.model_selection.cross_val_score(clf, X_train, y_train, cv=5, scoring="accuracy")
    print("Train CV Accuracy: %0.3f (+/- %0.3f) [%s]" % (scores.mean(), scores.std(), label))
    md = clf.fit(X_train, y_train)
    clfs.append(md)
    print("Test Accuracy: %0.4f " % (sklearn.metrics.accuracy_score(clf.predict(X_test), y_test)))
```

```python
XGB = XGBClassifier(n_estimators = 100, seed=666, eval_metric='logloss')
```
